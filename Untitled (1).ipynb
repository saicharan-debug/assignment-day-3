# question-2

```python
import numpy as np
x=[5,78,94,67,2]
y=[2,6,9,3,1]
arr_1=np.array(x)
arr_2=np.array(y)
print(arr_1,arr_2)
a=np.concatenate((arr_1,arr_2), axis=0)
print(a)
arr=np.sort((a), axis=0)
print(arr)
```
